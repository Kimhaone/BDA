```python
import pandas as pd
jeju_traff = pd.read_csv('jeju_traffic.csv')
```

```python
len(jeju_traff['road_name'].value_counts())
```

```python
jeju_sand = pd.read_csv('jeju_sand.csv', encoding = 'cp949')
```

```python
jeju_sand.columns = ['road_name', 'setting_name','location_count','per_count','setting_count','date']
```

```python
jeju_sand
```

```python
jeju_traff_ = jeju_traff[['road_name','maximum_speed_limit','target']]
```

```python
jeju_traff_
```

```python
#그룹화
group_traff = jeju_traff_.groupby(['road_name']).mean()
group_traff
```

```python
#병합
df1 = pd.merge(left = group_traff, right = jeju_sand, on = 'road_name', how = 'left') 
```

```python
df1
```

```python
df1['setting_count'].value_counts()
```

```python
df1.isna().sum()
```

```python
df_final = df1.dropna()
```

```python
df_final
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import rc
rc('font', family = 'Malgun Gothic')
```

```python
fig, ax= plt.subplots(nrows=1, ncols=2, figsize=(10,8))
ax[0].plot(df_final['road_name'], df_final['maximum_speed_limit'])#도로별 평균 최대 속도 
ax[1].bar(df_final['road_name'], df_final['setting_count']) #도로별 제설모래주머니 설치 갯수
```
