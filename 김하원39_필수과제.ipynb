```python
df_1 = df[['age','education','education-num','marital-status','occupation','race','sex','hours-per-week','continent','class']]
```

```python
df_1
```

```python
df_1.dropna(inplace = True)
```

```python
df_1_base = df_1[['education','education-num','marital-status','occupation','race','sex','hours-per-week','continent','class']]
```

```python
df_base = pd.get_dummies(df_1_base)
```

```python
df_base = df_base*1
```

```python
df_base
```

```python
df_base.columns
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df_base.drop('class',axis=1),df_base['class'], test_size=0.4, random_state=1)
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
model.fit(X_train, y_train)
pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, pred)
```
