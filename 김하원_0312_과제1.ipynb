```python
import pandas as pd
mpg = pd.read_csv('mpg.csv')
mpg.head()
```

```python
#1.자동차들의 도시연비와 고속도로 연비의 합은 얼마일까?
mpg['ch_sum']=(mpg['cty']+mpg['hwy'])
mpg
```

```python
#2.도시연비와 고속도로 연비의 합이 큰 자동차 모델은 무엇을까?
mpg.sort_values('ch_sum', ascending = False)
```

```python
#3.자동차들의 도시연비와 고속도로 연비의 평균은 얼마일까?
mpg['ch_mean']=((mpg['cty']+mpg['hwy'])/2)
mpg
```

```python
#4.구동방식 별 자동차 배기량의 평균은 어떻게 될까?
mpg_1= mpg.groupby('drv').agg(displ_mean = ('displ','mean'))
mpg_1
```

```python
#5. 이를 막대그래프로 표현해보자.
import matplotlib.pyplot as plt
plt.bar(['4','f','r'], [3.998058,2.558491,5.176000])
plt.show()
```

```python
#6.자동차 제조년도가 최근일 수록 도시와 고속도로 연비의 평균이 높을까?

mpg.head()
mpg1=mpg[['year','ch_mean']]
mpg1.groupby('year').agg(year_mean = ('ch_mean','mean'))
```

```python
#7.자동차의 평균 배기량 보다 높은 자동차는 몇개일까?
displ_mean = mpg['displ'].mean()
displ_mean

import numpy as np

mpg.assign(test = np.where(mpg['displ']>displ_mean, 'high','low'))
```

```python
#8.자동차 내 실린더 수 중에 가장 많은 비중을 차지하는 것은 몇개인가?
mpg.groupby('cyl').agg(n = ('model','count'))
```

```python
#9.가장 높은 통합연비를 지닌 자동차 종류는 무엇인가?
ch_mean_category = mpg.groupby(['category']).agg(ch_mean_cate = ('ch_mean','max'))
ch_mean_category.sort_values('ch_mean_cate', ascending = False)
```

```python
#10.자동차들 중 고속도로 연비가 도시 연비보다 더 큰 자동차의 갯수는 몇개인가?

mpg2 = mpg[['cty','hwy']]
mpg3= mpg2.assign(test = np.where(mpg2['hwy']>mpg2['cty'], 'high','low'))
mpg3['test'].value_counts()
```
