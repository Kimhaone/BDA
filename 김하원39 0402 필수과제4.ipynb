```python
import numpy as np
import matplotlib.pyplot as plt

x = (np.random.random(100)-0.5)*4
y = (np.random.random(100)-0.5)*4
```

```python
plt.scatter(x,y)
plt.show()
```

```python
#산점도 색, 투명도 지정
plt.scatter(x,y, color = 'red', alpha = 0.4)
```

```python
#점 크기 조정
plt.scatter(x,y, s = 150)
```

```python
#마커 모양 변경
plt.scatter(x,y, marker='x')
```

```python
#산점도 여러게 겹치기, 범례 설정

x1 = (np.random.random(100)-0.5)*4
y1 = (np.random.random(100)-0.5)*4

plt.scatter(x, y, color = 'purple', alpha = 0.6, label = 'uniform')
plt.scatter(x1,y1, color = 'orange', alpha = 0.7, label = 'normal')

plt.legend(loc = 'upper left')
plt.show()
```

```python
#seraborn 활용, 사이즈 조절

import seaborn as sns
sns.set(style = 'whitegrid')
plt.scatter(x,y)
plt.figure(figsize = (10,5))
plt.show()
```
