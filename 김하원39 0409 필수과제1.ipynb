## 도수분포표 란?

주어진 자료를 몇 개의 계급으로 나누고 각 계급의 도수를 조사하여 나타낸 표를 도수분포표라고 한다. 이때 변량을 일정한 간격으로 나눈 구간을 계급, 각 계급에 속하는 자료의 수를 도수라고 한다.

```python
#랜덤으로 히스토그램 생성
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

x = np.random.normal(5,10,100)
plt.hist(x, bins = 5)
```

```python
#한글 사용
plt.rc('font', family = 'NanumGothic')

#그래프 크기
fig = plt.figure(figsize = (8,6))

plt.hist(x, bins = 5)

#x축 ticks
plt.xticks(range(5), ['a','b','c','d','e'])
plt.tick_params(axis = 'x', direction='in', labelsize = 12, pad = 20)

#제목, x라벨, y라벨 지정
plt.title('히스토그램', fontsize = 30, pad = 30)
plt.xlabel('x축', fontsize = 20, loc = 'center', labelpad = 20)
plt.ylabel('y축', fontsize = 20, rotation = 0, loc = 'center', labelpad = 30)

plt.show()
```
