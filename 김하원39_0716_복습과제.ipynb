```python
import numpy as np
cos_sim = []
for i in range(len(all_words_df1)):
    cos_text = all_words_df1.iloc[i]
    cos = np.dot(target_text, cos_text) /(np.linalg.norm(target_text)*np.linalg.norm(cos_text))
    cos_sim.append(cos)

all_words_df1['cos_sim'] = cos_sim
```
