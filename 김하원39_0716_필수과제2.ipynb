```python
import pandas as pd
review = pd.read_csv('movie_rv.csv')
review
```

```python
review1 = review[review['label'] == 1]
review1= review1.dropna()
review0 = review[review['label'] == 0]
review0= review0.dropna()
```

```python
rv1_s = review1.sample(n=1500, random_state = 42)
rv0_s = review0.sample(n=1500, random_state = 42)

rv1_s
```

```python
from konlpy.tag import Okt 
from konlpy.tag import Twitter 

twt = Okt()
```

```python
parts =['Noun']
all_words_df1 = pd.DataFrame()
rv1_s_doc = []
for n in range(len(rv1_s)):
    text = rv1_s['document'].iloc[n]
    words = twt.pos(text)
    words_df = pd.DataFrame()
    
    for i in words:
        if i =='EOS' or i =='':continue
        word_tmp = i[0]
        part = i[1]
        if not (part in parts):continue
        words_df[word_tmp]=[1] 
    all_words_df1 = pd.concat([all_words_df1, words_df],ignore_index =True)
```

```python
all_words_df1
```

```python
all_words_df1 = all_words_df1.fillna(0)
all_words_df1
```

```python
parts =['Noun']
all_words_df0 = pd.DataFrame()
rv0_s_doc = []
for n in range(len(rv1_s)):
    text = rv0_s['document'].iloc[n]
    words = twt.pos(text)
    words_df = pd.DataFrame()
    
    for i in words:
        if i =='EOS' or i =='':continue
        word_tmp = i[0]
        part = i[1]
        if not (part in parts):continue
        words_df[word_tmp]=[1] 
    all_words_df0 = pd.concat([all_words_df0, words_df],ignore_index =True)
```

```python
all_words_df0
```

```python
all_words_df0 = all_words_df0.fillna(0)
all_words_df0
```

```python
target1 = all_words_df1.iloc[1]
target0 = all_words_df0.iloc[1]
```

```python
import numpy as np
cos_sim = []
for i in range(len(all_words_df1)):
    cos_text = all_words_df1.iloc[i]
    cos = np.dot(target1, cos_text) /(np.linalg.norm(target1)*np.linalg.norm(cos_text))
    cos_sim.append(cos)

all_words_df1['cos_sim'] = cos_sim
```
