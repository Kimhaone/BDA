```python
import pandas as pd
import numpy as np
```

```python
mpg = pd.read_csv('mpg.csv')
```

```python
mpg
```

```python
#1. append: 새로운 행 추가
df = pd.DataFrame(data=[['volkswagen','passat',3.6,2008,6,'auto(s6)','f',17,26,'p','midsize']],
                   columns=['manufacturer', 'model', 'displ', 'year', 'cyl', 'trans', 'drv', 'cty',
       'hwy', 'fl', 'category'])
```

```python
mpg.append(df, ignore_index= 234)
```

```python
#2. descirbe: 수치형데이터의 통계량
mpg.describe()
```

```python
#3. drop_duplicates: 중복행 제거
mpg.drop_duplicates()
```

```python
#4.get_values: 시리즈 값 구하기, 오류는 해당 메써드가 데이터프레임엔 존재하지 않기때문
mpg.get_values() 
```

```python
#5.isin: 특정 요소가 있는지 확인하는 함수, bool타입으로 결과 반환
mpg['category'].isin(['midsize'])
```

```python
#6.mean: 평균값 구하기
mpg['cty'].mean()
```

```python
#7.equals: 일치 확인
df1 = df
df2 = df.iloc[:, 3:]
```

```python
df1.equals(df)
```

```python
df2.equals(df)
```

```python
#8.sort_values(): 정렬하기, 특정 기준이나 오름/내림차순으로 정렬가능
mpg.sort_values(by = 'year')
```

```python
mpg.sort_values(by = 'year', ascending=False)
```

```python
#9.to_frame(): 시리즈를 데이터프레임으로 변환
a = pd.Series(['r','p','e','d','c'])
a.to_frame()
```
